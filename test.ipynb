```python
import matplotlib.pyplot as plt

years = [2000, 2005, 2010, 2015, 2020]
population = [6.1, 6.5, 6.9, 7.3, 7.8]

fig, ax = plt.subplots(figsize=(4, 2))

ax.plot(years, population, 'r--o', linewidth=2, markersize=10)
ax.set_title('Population Growth (2000-2020)', fontsize=18, color='green')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Population (Millions)', fontsize=12)

# Add informative tick labels
ax.set_xticks(years)
ax.set_xticklabels(['{:.0f}'.format(y) for y in years])

ax.grid(True, linestyle=':', color='lightgray')
ax.legend(['Population'], loc='upper left', frameon=False)

plt.savefig('good_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

countries = ['USA', 'India', 'China', 'Brazil', 'Germany']
gdp = [21.4, 2.9, 14.3, 1.8, 3.8]
inflation = [2.3, 6.2, 2.4, 3.7, 1.4]

x = np.arange(len(countries))

plt.figure(figsize=(6, 6))
plt.bar(x, gdp, color='lightblue', label='GDP')
plt.plot(x, inflation, color='black', linestyle='--', label='Inflation')
plt.xticks([0, 1, 2, 3, 4], ['USA', 'India', 'China', 'Brazil', 'Germany'])
plt.yticks(np.arange(0, 25, 5), ['0', '5', '10', '15', '20'])
plt.title('ECONOMICS DATA!!!')
plt.xlabel('Country')
plt.ylabel('Value (USD/Year)')
plt.legend(loc='upper right')
plt.grid(True)
plt.savefig('good_plot_2.png')
plt.show()
```
